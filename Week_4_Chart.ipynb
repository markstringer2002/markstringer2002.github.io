First we import pandas

```python
import pandas as pd
```

We read our html file that we accessed in the elements of the ESPN page

```python
table_list=pd.read_html('olympics.html')
```

Seeing the variables in our table, how they are named and how they are coded

```python
table_list[0].info()
```

Converting this to a dictionary

```python
olympics_df=table_list[0]
```

```python
olympics_df
```

Seeing how the columns are labelled

```python
olympics_df.columns
```

Renaming the columns to be more descriptive, and redefining our dictionary name

```python
medals_df = olympics_df.rename(columns={'Group':'Country', 'G':'Gold','S':'Silver', 'B':'Bronze' })
```

```python
medals_df
```

We only want the 10 countries with the highest medal counts, so the bottom 75 observations were removed

```python
n = 75
medals_df1 = medals_df.head(-n)
```

```python
medals_df1
```

We do not want the column 'Total' to display as part of the stacked bar chart, so we remove this element

```python
medals_df1=medals_df1.drop('Total', axis=1)
```

```python
medals_df1
```

We want to convert this into tidy data form, so stack by country to get 3 rows for each country, one for gold, silver and bronze

```python
medals_df2=medals_df1.set_index('Country').stack().reset_index()
```

```python
medals_df2
```

We need to change the column names to something we understand

```python
medals_df3 = medals_df2.rename(columns={'level_1':'Medal Type', 0:'Value', })
```

```python
medals_df3
```

We save this cleaned data to a csv to be able to link to Vega

```python
medals_df3.to_csv('medalsnew_data.csv')
```
